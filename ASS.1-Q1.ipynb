```python
import yfinance as yf 
import numpy as np
import matplotlib.pyplot as plt
```

GRAPH PLOTTING

```python
def graph(ticker, start="2023-12-10",end="2024-12-10", interval="1d"):
    data=yf.download(tickers=ticker,start=start,end=end)
    close=data["Close"]
    print("AAPL stock price from 2023-12-10 to 2024-12-10:" )
    plt.plot(close)
    
    plt.xlabel("DATE",size="15")
    plt.grid()
    plt.ylabel("Closing price in USD",size="15")
    plt.title("AAPL",size="20")
    plt.show()
graph("AAPL")
print("""The graph indicates a bearish trend for the first 5 months, however we  then see a bullish trend as prices soar up till \n
the very end of the period. A relative fall in the price of the stock is seen in the month of August which has recovered to \n
reach the highest price last year.""")
```
